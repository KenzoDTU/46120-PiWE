```python
#only takes integer

def goldilocks(x):
    if x < 140:
        print("Too small!")
    elif x > 150:
        print("Too big!")
    else:
        print("Just right!")
```

```python
def square_list(L):
    output = []
    for x in L:
        output.append(x**2)
    return output
```

```python
def fibonacci_stop(n):
    a, b = 0, 1
    output = []
    while a < n:
        output.append(a)
        a, b = b, a + b
    return output
```

```python
def clean_pitch(x, status):
    output = []
    for xval, stat in zip(x, status): 
        if stat == 0:
            output.append(xval)
        elif stat > 0:
            output.append(-999)
    return output
```
